```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
#
import matplotlib.pyplot as plt

# 1.figure
fig = plt.figure()

# 2.1.figureaxes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.axes
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.
plt.show()
```

```python
figure = plt.figure()
ax1 = figure.add_subplot(1, 3, 1)
ax1.set_title("Axes1")
ax2 = figure.add_subplot(1, 3, 2)
ax2.set_title("Axes2")
ax3 = figure.add_subplot(1, 3, 3)
ax3.set_title("Axes3")

plt.show()
```

```python
figure = plt.figure(figsize=(4,3),dpi=180)
student = [80, 92, 75, 89, 69]
axes = figure.add_subplot()
axes.plot(['chinese', 'english', 'math', 'history', 'created'],student)
plt.show()
```

```python
figure = plt.figure(figsize=(8,5),dpi=120)
stu1 = {'chinese':80, 'english':92, 'math':75, 'history':89, 'created':69}
stu2 = {'chinese':70, 'english':83, 'math':60, 'history':92, 'created':77}
axes = figure.add_subplot()
axes.plot(stu1.keys(),stu1.values())
axes.plot(stu2.keys(),stu2.values())
plt.show()
```
